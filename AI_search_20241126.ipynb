```python
import gradio as gr
import requests

def get_history_messages(histories):
    history_list = list()

    history_length = 5
    history_index = 0

    for history in histories:
        
        if history_index >= history_length:
            break
        
        message1 = history[0]
        message2 = history[1]
        
        history_list.append({
            "role": "assistant",
            "content": message1
        })
        history_list.append({
            "role": "assistant",
            "content": message2               
        })
        
        history_index += 1
        
    return history_list


def request_gpt(prompt, history_list):
    endpoint = "your endpoint"
    # post
    
    ai_search_endpoint = "your endpoint"
    ai_search_key = "your key"
    ai_search_index = "theme-index"
    
    headers = {
        "Content-Type": "application/json",
        "api-key": "your key"
    }
    message_list = list()
    
    # system
    message_list.append({
        "role": "system",
        "content": "사용자가 정보를 찾는 데 도움이 되는 AI 도우미입니다."       
    })
    # 이전 메시지
    message_list.extend(history_list)
    # prompt
    message_list.append({
        "role": "user",
        "content":prompt
    })
    
    payload = {
        "messages": message_list,
        "temperature": 0.7,
        "top_p": 0.95,
        "max_tokens": 800,
        "data_sources": [
        {
            "type": "azure_search",
            "parameters": {
                "endpoint": ai_search_endpoint,
                "semantic_configuration": "theme-semantic",
                "query_type": "semantic",
                "strictness": 3,
                "top_n_documents": 5,
                "key": ai_search_key,
                "indexName": ai_search_index
            }
        }
        ]
    }
    
    response = requests.post(endpoint, headers=headers, json=payload)
    
    if response.status_code == 200:
        response_json = response.json()
        content = response_json["choices"][0]["message"]["content"]
        return content
    else:
        return f"{response.status_code}, {response.text}"


def click_send(prompt, histories):
    print(prompt)
    print(histories)
    history_list = get_history_messages(histories=histories)
    response_text = request_gpt(prompt, history_list)
    histories.append((prompt, response_text))
    return histories, ""


# 스타일을 추가하는 부분
css = """
#chatbot {
    background-color: #E6E6FA; /* 연보라색 */
    border-radius: 10px;
    padding: 10px;
    height: 400px;
    overflow-y: scroll;
}

#input_textbox {
    border-radius: 15px;
    padding: 10px;
    font-size: 16px;
    background-color: #f9f9f9;
}

#send_button {
    background-color: #6A5ACD; /* 중간 보라색 */
    color: white;
    border-radius: 20px;
    padding: 10px 20px;
    font-size: 16px;
}

#send_button:hover {
    background-color: #483D8B; /* 어두운 보라색 */
}
"""

with gr.Blocks(css=css) as demo:
    chatbot = gr.Chatbot(label="채팅 기록", elem_id="chatbot")
    
    with gr.Row():
        input_textbox = gr.Textbox(label="", scale=7, elem_id="input_textbox")
        send_button = gr.Button("전송", scale=1, elem_id="send_button")
    
    input_textbox.submit(fn=click_send, inputs=[input_textbox, chatbot], outputs=[chatbot, input_textbox])
    send_button.click(fn=click_send, inputs=[input_textbox, chatbot], outputs=[chatbot, input_textbox])
    
demo.launch()
```
